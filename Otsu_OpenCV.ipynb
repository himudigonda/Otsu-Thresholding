# Imports

```python
import cv2
import matplotlib.pyplot as plt
```

# Read Images

```python
images = ['./assets/cat.jpg','./assets/dog.jpg', './assets/hamster.jpg', './assets/otter.jpeg']
```

```python
image = cv2.imread(images[0], cv2.IMREAD_GRAYSCALE)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

for i, image_path in enumerate(images):
    row = i //2
    col = i % 2
    image = cv2.imread(image_path)
    if image is not None:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        axs[row, col].imshow(image_rgb)
        # axs[row, col].axis('off')

plt.tight_layout()
plt.show()
```

# OTSU Thresholding

```python
o_images = images
t_images = []

def otsu(image):
    if image is not None:
        _, thresholded_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        t_images.append(thresholded_image)

```

```python
for i in o_images:
    image = cv2.imread(i, cv2.IMREAD_GRAYSCALE)
    otsu(image)
```

```python
len(t_images)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Loop through the arrays and display them in subplots
for i, img_array in enumerate(t_images):
    row = i // 2
    col = i % 2

    # Display the array in the appropriate subplot
    axs[row, col].imshow(img_array, cmap='gray')  # Use 'cmap' for grayscale images
    # axs[row, col].axis('off')  # Hide axes

# Adjust the layout to prevent overlapping
plt.tight_layout()

# Show the combined figure with all four arrays
plt.show()





```
